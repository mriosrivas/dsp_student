## Part 4: Test your `fir.py` file

## 4.1 Test your window filter

```python
fc = 0.20
M = 103
L = 1024

h = flt.window_filter(fc, M, window_type='blackman', normalized=True)
g = flt.window_filter(fc, M, window_type='hamming', normalized=True)

h_inverted = flt.spectral_inversion(h)
g_inverted = flt.spectral_inversion(g)

h_reversal = flt.spectral_reversal(h)
g_reversal = flt.spectral_reversal(g)
```

```python
# Spectral inversion 
H, f = auxiliary.zero_pad_fourier(h, M)
G, fg = auxiliary.zero_pad_fourier(g, M)
H_inv, f_h_inv = auxiliary.zero_pad_fourier(h_inverted, M)
H_rev, f_h_rev = auxiliary.zero_pad_fourier(h_reversal, M)
```

```python
plt.rcParams["figure.figsize"] = (10, 10)

plt.subplot(2,2,1)
cplots.plot_frequency_response(np.absolute(H), f, title="Inversion Blackman")
plt.subplot(2,2,2)
cplots.plot_frequency_response(np.absolute(G), fg, title="Inversion Hamming")
plt.subplot(2,2,3)
cplots.plot_frequency_response(np.absolute(H_inv), f_h_inv, title="Reversal Blackman")
plt.subplot(2,2,4)
cplots.plot_frequency_response(np.absolute(H_rev), f_h_rev, title="Reversal Hamming")
plt.subplots_adjust(hspace=0.3)
```

### 4.2 Test your low pass filter

```python
fc = 0.2
M = 101
h_lp = flt.low_pass_filter(fc, M)
H_lp, f_lp = auxiliary.zero_pad_fourier(h_lp, M)
```

```python
plt.rcParams["figure.figsize"] = (5, 4)
cplots.plot_frequency_response(np.absolute(H_lp), f_lp)
plt.grid("on")
```

### 4.3 Test your high pass filter

```python
h_hp = flt.high_pass_filter(fc, M, method='spectral_inversion')
H_hp, f_hp = auxiliary.zero_pad_fourier(h_hp, M)
```

```python
g_hp = flt.high_pass_filter(fc, M, method='spectral_reversal')
G_hp, g_hp = auxiliary.zero_pad_fourier(g_hp, M)
```

```python
cplots.plot_frequency_response(np.absolute(H_hp), f_hp, label='spectral inversion')
cplots.plot_frequency_response(np.absolute(G_hp), g_hp, label='spectral reversal')
plt.legend()
plt.grid("on")
```

### 4.2 Test your band/reject band filter

```python
fc1 = 0.1
fc2 = 0.2

h = flt.band_filter(fc1, fc2, M, band_type='pass')
H, f = auxiliary.zero_pad_fourier(h, M)

g = flt.band_filter(fc1, fc2, M, band_type='reject')
G, f = auxiliary.zero_pad_fourier(g, M)
```

```python
cplots.plot_frequency_response(np.absolute(H), f, label='Band Pass Filter')
cplots.plot_frequency_response(np.absolute(G), f, label='Band Reject Filter')
plt.legend()
```
